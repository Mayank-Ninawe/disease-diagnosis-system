```python
# Disease Diagnosis System - Data Exploration
# Author: [Your Name]
# Date: October 19, 2025

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data (relative path)
df = pd.read_csv('../data/Training.csv')

print("Dataset Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
# Dataset information
print("Dataset Info:")
print(f"Total rows: {len(df)}")
print(f"Total columns: {len(df.columns)}")
print(f"\nColumn names:\n{df.columns.tolist()}")

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum().sum())

# Disease distribution
print("\nNumber of unique diseases:", df['prognosis'].nunique())
print("\nDisease list:")
print(df['prognosis'].unique())
```

```python
# Disease distribution bar chart
plt.figure(figsize=(12, 6))
df['prognosis'].value_counts().plot(kind='bar')
plt.title('Disease Distribution in Dataset')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('disease_distribution.png')
plt.show()

print("âœ… Data Exploration Complete!")
```
